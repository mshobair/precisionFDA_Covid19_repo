### Read the matrices and save in memory


```r
# generate string vector with name of the matrix files
motif_matrix_name_list = {}

motif_matrix_list = {}

for(subject_ind in 2:200){
    motif_matrix_name_list[[subject_ind]] <- paste0("PCM2_", subject_ind)
    motif_matrix_list[[subject_ind]] <- paste0("PCM2_motif2_", subject_ind)

    
}
```

```r
PCM_table2 = {}
PCM_table2_col = {}
PCM_table2_col_mat = {}

for(subject_ind in 15:150){ # looping through subject_id indices
    # matrix 
    PCM_table2[[subject_ind]] <- read.table(paste0("~/Documents/Projects/", 
                     motif_matrix_name_list[[subject_ind]]), 
                         header = FALSE,
                         row.names = 1)
    # col name
   
    PCM_table2_col[[subject_ind]] <- read_yaml(paste0("~/Documents/Projects/", 
                     motif_matrix_list[[subject_ind]]))[8]
    
    names(PCM_table2[[subject_ind]]) <- strsplit(PCM_table2_col[[subject_ind]][[1]],
                                                 split = "")[[1]]
    # make matrix
    PCM_table2_col_mat[[subject_ind]] <- as.matrix(
                                            PCM_table2[[subject_ind]])
    names(PCM_table2_col_mat)[[subject_ind]] <- unique(metadata_table$subject_id)[subject_ind]
        }
```

```r
tree <-motif_tree(PCM_table2_col_mat[15:150], layout = "daylight"
                )
tree
```

```r
# function to plot motif by subject_id
plot_motif <- function(pcm){
protein<-pcm2pfm( pcm )
protein_motif<-new("pfm", mat=protein, name="CAP", 
                   color=colorset(alphabet="AA",colorScheme="chemistry"))

plot(protein_motif)
}
```
